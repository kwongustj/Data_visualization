```python
import pandas as pd  
import matplotlib as mpl       
import matplotlib.pyplot as plt   

pd.set_option('display.max_row', 1000)
pd.set_option('display.max_columns', 10)
 
df = pd.read_csv('roadkill.csv', index_col='노선')
df  
```

```python
df1=df.groupby(['연도','노선']).describe()
df1
```

```python
df3=df.groupby(['노선','연도']).describe()
df3
```

```python
df7=pd.read_csv('roadkill.csv',sep=',')
plt.rc('font', family='NanumGothic') # For Windows
animal=df7['동물종류'].value_counts()
animal.plot(kind = 'bar')
```
